```python
_list = np.concatenate([np.array(i) for i in data['production_companies'].str.split('|')])
production_companies = list(set(_list))

#production_companies=['Four By Two Productions']

_dict = {}
for comp in production_companies:    
  compMovies = data[data['production_companies'].str.contains(comp)]
  titles = compMovies['original_title'] 
  arr=[]
  for title in titles: 
    arr.append(len(title))    
  npArray = np.array(arr)
  mean=0;
  if len(npArray)>0:
    mean = npArray.mean()
  _dict[comp] = mean;

cc = Counter(_dict)
print(cc)

  
#{'Four By Two Productions': 83.0}
answer_ls.append(5)
```

# 35. Какие актеры чаще всего снимаются в одном фильме вместе
Варианты ответа:
1. Johnny Depp & Helena Bonham Carter
2. Hugh Jackman & Ian McKellen
3. Vin Diesel & Paul Walker
4. Adam Sandler & Kevin James
5. Daniel Radcliffe & Rupert Grint

```python
_list = np.concatenate([np.array(i) for i in data['cast'].str.split('|')])
actors = list(set(_list))

#actors=["Vin Diesel"]

arr=[]
for actor in actors:  
  otherActors = data[data['cast'].str.contains(actor)]['cast']
  for others in otherActors.values:
    masOthersActors = others.split('|')
    for otherActor in masOthersActors:
      if(otherActor!=actor):
        key=actor+"_"+otherActor
        arr.append(key)

cc = Counter(arr)
print(cc)
      
#'Rupert Grint_Daniel Radcliffe': 8

answer_ls.append(5)
```

# 36. У какого из режиссеров выше вероятность выпустить фильм в прибыли? (5 баллов)

```python
#отношение всех фильмов к прибыльным
#кол-во всех, кол-во прибылных. сделать отд. столбец, отсортировать

directors = data['director'].value_counts().to_frame()
directors = directors.reset_index()
directors.columns=["dir","total"]
#display(directors)

incomePlusDirectors = data[data.income>0]['director'].value_counts().to_frame()
incomePlusDirectors = incomePlusDirectors.reset_index()
incomePlusDirectors.columns=["dir","plus"]
#display(incomePlusDirectors)

joined = incomePlusDirectors.merge(directors, on='dir', how='left')
joined["rez"] = joined["plus"]/joined["total"]
joined=joined[joined["rez"]==1]
display(joined)
answer_ls.append(2)
```
